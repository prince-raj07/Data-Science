### Pandas Series

A Pandas Series is like a column in a table. It is a 1-D array holding data of any type.

```python
import numpy as np 
import pandas as pd 


```

###Series from lists

```python
#string

country = ['India','Pakistan','USA','Nepal']
pd.Series(country)
```

```python
#integers
runs = [12,24,56,78,100]

runs = pd.Series(runs)
runs
```

```python
#custom index

marks = [67,57,89,100]
subjects = ['maths','english','science','hindi']

pd.Series(marks,index=subjects)
```

```python
marks = pd.Series(marks,index=subjects,name='Nistish ke marks')
marks
```

###Series from dict

```python
marks  = {
    'maths' :67,
    'english':57,
    'science' :89,
    'hindi' :100
}

marks_series= pd.Series(marks,name='nitish ke marks')
marks_series
```

### Series Maths Methods
